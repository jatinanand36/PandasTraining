```python
import pandas as pd
import numpy as np
N=20
DF = pd.DataFrame({
    'A':pd.date_range(start='2019-06-01',periods=N,freq='D'),
    'x':np.linspace(0,stop=N-1,num=N),
    'y':np.random.rand(N),
    'C':np.random.choice(['Low','Medium','High'],N).tolist(),
    'D':np.random.normal(100,10,size=(N)).tolist()
})
```

```python
DF
```

```python
for col in DF:
    print(col)
```

```python
# iteritems()
for key,val in DF.iteritems():
    print(key,val)
```

```python
# iterrows
for row_index,row in DF.iterrows():
    print(row_index,row)
```

```python
# itertuples()
for row in DF.itertuples():
    print(row)
```

```python
for index,row in DF.iterrows():
    print(row['A'])
    
```
